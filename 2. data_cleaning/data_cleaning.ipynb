```python
import re
import xlrd
import pyvi
import numpy as np
import pandas as pd
from pyvi import ViTokenizer

#đường dẫn file dữ liệu
loc = ("C:/Users/DELL/(K) Khanh_graduates_excellently/2. data_cleaning/raw_dataset.xlsx")
#đọc file excel 
raw_data = xlrd.open_workbook(loc)
```

```python
#lấy sheet 1 của database
train_sheet = raw_data.sheet_by_index(0)
#lấy sheet 2 của database
test_sheet = raw_data.sheet_by_index(1)

#đưa dữ liệu sheet 1 vào trong một list
train_list1 = []
for i in range(train_sheet.nrows):
    train_list1.append(train_sheet.cell_value(i, 0))
#đưa dữ liệu sheet 2 vào trong một list
test_list1 = []
for i in range(test_sheet.nrows):
    test_list1.append(test_sheet.cell_value(i, 0))

#bóc tách reviews và scores
train_list2 = []
for x in train_list1:
    for y in x.split("< > *"):
        train_list2.append(y)
train_list3 = []
train_part_list = []
for i in range(len(train_list2)):
    if i%2 == 0:
        train_list3.append(train_list2[i])
    if i%2 != 0:
        train_part_list.append(train_list2[i])
train_list4 = []
for x in train_list3:
    for y in x.split(" < > "):
        train_list4.append(y)
train_part_list_scores = []
for i in range(len(train_list4)):
    if i%2 != 0:
        train_part_list_scores.append(train_list4[i])
    if i%2 == 0:
        pass

test_list2 = []
for x in test_list1:
    for y in x.split("< > *"):
        test_list2.append(y)
test_list3 = []
test_part_list = []
for i in range(len(test_list2)):
    if i%2 == 0:
        test_list3.append(test_list2[i])
    if i%2 != 0:
        test_part_list.append(test_list2[i])
test_list4 = []
for x in test_list3:
    for y in x.split(" < > "):
        test_list4.append(y)
test_part_list_scores = []
for i in range(len(test_list4)):
    if i%2 != 0:
        test_part_list_scores.append(test_list4[i])
    if i%2 == 0:
        pass    

print(train_part_list_scores)
print(len(train_part_list_scores))
print(train_part_list)
print(len(train_part_list))
print(" ")
print(" ")
print(" ")
print(test_part_list_scores)
print(len(test_part_list_scores))
print(test_part_list)
print(len(test_part_list))
```

```python
#Làm sạch ký tự dấu xuống dòng, và các ký tự lạ khác
space_replacing1 = re.compile("[^a-zA-Z_0-9ÀÁÂÃÈÉÊÌÍÒÓÔÕÙÚĂĐĨŨƠàáâãèéêìíòóôõùúăđĩũơƯĂẠẢẤẦẨẪẬẮẰẲẴẶẸẺẼỀỀỂưăạảấầẩẫậắằẳẵặẹẻẽềềểỄỆỈỊỌỎỐỒỔỖỘỚỜỞỠỢỤỦỨỪễệỉịọỏốồổỗộớờởỡợụủứừỬỮỰỲỴÝỶỸửữựỳỵỷỹế\\s]")
space_replacing2 = re.compile("(<br\s*/><br\s*/>)|(\-)|(\/)|(\n)")

def space_cleaning (reviews):
    cleaned_reviews = []
    for review in reviews:
        review = space_replacing1.sub(" ", str(review).lower())
        review = space_replacing2.sub(" ", str(review))
        review = " ".join(review.split())
        cleaned_reviews.append(review)
    return cleaned_reviews

train_clean = space_cleaning(train_part_list)
test_clean = space_cleaning(test_part_list)

print(train_clean)
print(test_clean)
print(len(train_clean))
print(len(test_clean))
```

```python
#tokenizing
def review_tokenizing (reviews):
    tokenized_reviews = []
    for review in reviews:
        tokenized_reviews.append(ViTokenizer.tokenize(review))
    return tokenized_reviews

tokenized_train = review_tokenizing(train_clean)
tokenized_test = review_tokenizing(test_clean)

print(tokenized_train)
print(tokenized_test)
print(len(tokenized_train))
print(len(tokenized_test))
```

```python
#đếm số từ giảm sau khi tokenize cho phần train
wordcount = 0
for review in train_clean:
    words = review.split(' ')
    for word in words:
        wordcount += 1       
print(wordcount)
wordcount = 0
for review in tokenized_train:
    words = review.split(' ')
    for word in words:
        wordcount += 1        
print(wordcount)
print(" ")
#đếm số từ giảm sau khi tokenize cho phần test
wordcount = 0
for review in test_clean:
    words = review.split(' ')
    for word in words:
        wordcount += 1       
print(wordcount)
wordcount = 0
for review in tokenized_test:
    words = review.split(' ')
    for word in words:
        wordcount += 1        
print(wordcount)
print(" ")


#thư viện stop words và số lượng từ trong thư viện
stop_words = []
stop_words = [line.strip('\n') for line in open('vietnamese-stopwords-dash.txt', 'r', encoding='utf-8')]
count_sw = 0 
for word in stop_words:
    count_sw +=1
print(count_sw)
print(stop_words)
```

```python
#loại bỏ stop words
def stop_words_removing (reviews):
    final = []
    for review in reviews: 
        removed = []
        for word in review.split(" "):
            if (word not in stop_words):
                removed.append(word)
        final.append(" ".join(removed))
    return final

final_train = stop_words_removing(tokenized_train)
final_test = stop_words_removing(tokenized_test)

print(final_train)
print(final_test)
print(len(final_train))
print(len(final_test))
```

```python
#đếm số từ giảm sau khi loại bỏ stop words cho phần train
wordcount = 0
for review in tokenized_train:
    words = review.split(' ')
    for word in words:
        wordcount += 1       
print(wordcount)
wordcount = 0
for review in final_train:
    words = review.split(' ')
    for word in words:
        wordcount += 1        
print(wordcount)
print(" ")
#đếm số từ giảm sau khi loại bỏ stop words cho phần test
wordcount = 0
for review in tokenized_test:
    words = review.split(' ')
    for word in words:
        wordcount += 1       
print(wordcount)
wordcount = 0
for review in final_test:
    words = review.split(' ')
    for word in words:
        wordcount += 1        
print(wordcount)
print(" ")
print("Xong phần tiền xử lý dữ liệu")
```
